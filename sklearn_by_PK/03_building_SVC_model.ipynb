Download wheat-dataset from above link.
We'll implement a model to predict the type of wheat.

```python
import pandas as pd
total_data=pd.read_csv("data_03/Seed_Data.csv")
total_data.describe
```

```python
X = total_data.iloc[:,:-1]
X
```

```python
y=total_data.iloc[:,-1]
y
```

```python
import sklearn
from sklearn import svm
from sklearn.svm import SVC
```

```python
# learning a model
from sklearn.model_selection import train_test_split
# by default the data is split in 3:1 ratio for train:test

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=13)
# test_size: test data = 20%
# random_state: any number (as a seed to random number generator)

# normalising data:
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
[X_train,X_test]
```

```python
clf=SVC()
clf.fit(X_train,y_train)

pred_clf = clf.predict(X_test)
sklearn.metrics.accuracy_score(y_test,pred_clf)
```

```python
print(sklearn.metrics.classification_report(y_test,pred_clf))
```
