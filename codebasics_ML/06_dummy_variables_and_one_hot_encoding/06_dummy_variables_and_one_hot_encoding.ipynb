```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

# Removing/replacing categorical variables with numerical using pandas.get_dummies()

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged_df = pd.concat([df,dummies], axis='columns')
merged_df = merged.drop(['town'],axis='columns')
```

# Drop one of the dummy variables to avoid 'dummy variable TRAP'
# Since, the dropped dummy variable can be derived from the remaining dummy variables
## P.S.: sklearn handles the 'dummy variable trap' even if we don't drop it. But it's a good practice to drop it from our side

```python
final_df = merged_df.drop(['west windsor'], axis='columns')
final_df
```

## Learn model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
X = final_df.drop('price',axis='columns')
y = final_df.price
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

## Check model accuracy using 'score'

```python
model.score(X,y)
```
