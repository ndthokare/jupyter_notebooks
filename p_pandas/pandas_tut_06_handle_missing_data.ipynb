# Handle missing data using "replace functions"

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data_t06/weather_data.csv")
df
```

```python
new_df = df.replace(-99999,np.NAN)
new_df
```

```python
new_df = df.replace([-99999,-88888],np.NAN)
new_df
```

```python
new_df = df.replace({
    "temperature":-99999,
    "windspeed":-88888,
    "event":"0"
},np.NAN)
new_df
```

```python
new_df = df.replace({
    -99999:np.NaN,
    -88888:np.NaN,
    "0":"Clear"
})
new_df
```

```python
# replace non-digit characters from temperature and windspeed columns
new_df = df.replace({
    "temperature":'[A-Za-z]',
    "windspeed":'[A-Za-z]',
},'',regex=True)
new_df
```
