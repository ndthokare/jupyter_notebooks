```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.plot(x,y,'g*-.')
```

```python
# D: Diamond
# plt.plot(x,y,color='red',marker='D',linestyle='')
plt.plot(x,y,'rD') # command using format string
# both the above commands give same output
```

```python
# plt.plot(x,y,color='red',marker='D',linestyle='-')
plt.plot(x,y,'rD-') # command using format string
# both the above commands give same output
```

```python
# Color in hex format
plt.plot(x,y,color='#FFFF00',marker='D', markersize=15,linestyle='-')
#plt.plot(x,y,'rD-') # command using format string
# both the above commands give same output
```

```python
plt.plot(x,y,color='green',alpha=0.5) # alpha range [0,1]. 0: Transperent, 1: Opaque
```
