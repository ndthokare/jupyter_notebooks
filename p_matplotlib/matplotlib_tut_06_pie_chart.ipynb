<h1 align="center", style="color:green">Matplotlib Tutorial: Pie Charts</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### **Draw pie chart to track down home expenses**

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
# plt.axes("equal") # this is required if the plot is not a perfect circle
plt.pie(exp_vals, labels=exp_labels)
```

# manage pie-chart size by 'radius'

```python
# plt.axes("equal") # this is required if the plot is not a perfect circle
plt.pie(exp_vals, labels=exp_labels, radius=2) # default radius=1
plt.show() # this removes the extra information displyed before the plot
```

# display the percentage using 'autopct'

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.0f%%") # default radius=1
plt.show() # this removes the extra information displyed before the plot
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%") # default radius=1
plt.show() # this removes the extra information displyed before the plot
```

# shadow parameter:


```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", shadow=True)
plt.show()
```

# explode to breakout some part of pie chart:

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", 
        shadow=True, explode=[0,0,0,0,0])
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", 
        shadow=True, explode=[0,.1,0,0.15,0])
plt.show()
```

# by default chart start at +x axis and draws clockwise change by 'startangle'

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", 
        shadow=True, explode=[0,.1,0,0.15,0], startangle=90)
plt.show()
```
