```python
xpos = np.arange(len(company)) # this is required since we cannot consider text as the value on axis to plot
plt.xticks(xpos,company) # this will assign text values inplace of the numbers
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue in $Bn")
plt.bar(xpos+0.2, profit, width=0.4, label="Profit in $Bn") # by default bars are drawn one above another due to same x-position
plt.legend()
```

# Horizontal bar charts using 'barh'

```python
ypos = np.arange(len(company))
plt.yticks(ypos,company) 

plt.ylabel("Company")
plt.xlabel("Revenue")
plt.title("US Tech Stocks")

plt.barh(ypos-0.2, revenue, height=0.4, label="Revenue in $Bn")
plt.barh(ypos+0.2, profit, height=0.4, label="Profit in $Bn")
plt.legend()
```
